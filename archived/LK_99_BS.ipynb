```python
def get_ibz_k_points():
  return fe_calc.get_ibz_k_points()
```

```python
fe_calc.get_number_of_spins()
```

```python
def get_number_of_spins():
  return fe_calc.get_number_of_spins()
```

```python
def get_eigenvalues(kpt, spin=0):
  return fe_calc.get_eigenvalues(kpt, spin)
```

```python
def get_fermi_level():
  return fe_calc.get_fermi_level()
```
